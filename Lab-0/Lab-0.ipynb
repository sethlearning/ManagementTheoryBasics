```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin
```

```python
def s(t):
    result = []
    for i in t:
        result.append(5*sin(i/2 + 1))
    return np.array(result)
```

```python
x = np.arange(0, 21, 0.1, dtype=float)
y = s(x)

xticks = range(22)
yticks = range(-5,6)
plt.plot(x, y)
plt.yticks(yticks)
plt.xticks(xticks)
plt.grid()


xd = range(22)
yd = s(xd)
plt.scatter(xd, yd, c='red', s=20)
plt.yticks(yticks)
plt.xticks(xticks)

plt.show()
```
